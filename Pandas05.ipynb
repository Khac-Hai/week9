##### Have you noticed that the type of Year is int64. But pandas has a different type to work with Time Series. Let's see it now.

### Step 5. Convert the type of the column Year to datetime64

```python
crime['Year'] = pd.to_datetime(crime['Year'], format='%Y')
print(crime.dtypes)  # kiểm tra lại kiểu dữ liệu
```

### Step 6. Set the Year column as the index of the dataframe

```python
crime = crime.set_index('Year')
crime.head()
```

### Step 7. Delete the Total column

```python
crime = crime.drop(columns=['Total'])
```

### Step 8. Group the year by decades and sum the values

#### Pay attention to the Population column number, summing this column is a mistake

```python
# Tạo cột decade
crime['Decade'] = (crime.index.year // 10) * 10

# Tính tổng theo thập kỷ, với Population lấy trung bình
grouped = crime.groupby('Decade').agg({
    'Population': 'mean',
    # giả sử các cột còn lại là các loại tội phạm, bạn có thể list hết các cột này,
    # hoặc tính tổng tất cả cột ngoại trừ Population
})

# Ví dụ nếu muốn tổng tất cả cột trừ Population
crime_sum = crime.drop(columns=['Population']).groupby('Decade').sum()

# Kết hợp Population lại
grouped = crime_sum.copy()
grouped['Population'] = crime.groupby('Decade')['Population'].mean()

print(grouped)
```

### Step 9. What is the most dangerous decade to live in the US?

```python
# Tổng các loại tội phạm (giả sử tất cả cột trừ Population)
grouped['TotalCrime'] = grouped.drop(columns=['Population']).sum(axis=1)

most_dangerous_decade = grouped['TotalCrime'].idxmax()
print("Most dangerous decade:", most_dangerous_decade)
```
