### Step 1. Import the necessary libraries:

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/tips.csv).

### Step 3. Assign it to a variable called tips

```python
tips = pd.read_csv('/content/tips.csv')
```

### Step 4. Delete the Unnamed 0 column

```python
if 'Unnamed: 0' in tips.columns:
    tips = tips.drop(columns=['Unnamed: 0'])
```

### Step 5. Plot the total_bill column histogram

```python
plt.figure(figsize=(8,5))
tips['total_bill'].hist(bins=30)
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.title('Histogram of Total Bill')
plt.show()
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.title('Scatter plot of Total Bill vs Tip')
plt.show()
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=tips, x='total_bill', y='tip', size='size', sizes=(20, 200), alpha=0.6)
plt.title('Total Bill vs Tip with Size')
plt.show()
```

### Step 8. Present the relationship between days and total_bill value

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=tips, x='day', y='total_bill')
plt.title('Boxplot of Total Bill by Day')
plt.show()
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=tips, x='tip', y='day', hue='sex', s=100)
plt.title('Scatter plot of Tip vs Day by Sex')
plt.show()
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=tips, x='day', y='total_bill', hue='time')
plt.title('Total Bill per Day by Time (Dinner/Lunch)')
plt.show()
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
plt.figure(figsize=(12,5))
sns.histplot(data=tips[tips['time']=='Dinner'], x='tip', color='blue', label='Dinner', kde=False, bins=20)
sns.histplot(data=tips[tips['time']=='Lunch'], x='tip', color='orange', label='Lunch', kde=False, bins=20)
plt.legend()
plt.title('Histogram of Tip for Dinner and Lunch')
plt.show()
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
fig, axes = plt.subplots(1, 2, figsize=(14,5), sharey=True)

sns.scatterplot(data=tips[tips['sex']=='Male'], x='total_bill', y='tip', hue='smoker', ax=axes[0])
axes[0].set_title('Male: Total Bill vs Tip by Smoker')
axes[0].legend(title='Smoker')

sns.scatterplot(data=tips[tips['sex']=='Female'], x='total_bill', y='tip', hue='smoker', ax=axes[1])
axes[1].set_title('Female: Total Bill vs Tip by Smoker')
axes[1].legend(title='Smoker')

plt.tight_layout()
plt.show()
```
