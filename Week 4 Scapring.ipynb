```python
import requests
from bs4 import BeautifulSoup

url= ["https://vietnamnet.vn/chinh-tri","https://vietnamnet.vn/chinh-tri-page1"]
list_link = []
list_href = []
for i in range(len(url)):
  response = requests.get(url[i])
  if response.status_code == 200:
    html_content = response.text
    soup = BeautifulSoup(html_content, 'html.parser')
    #print(soup.prettify())
    divs = soup.find_all('div', class_="horizontalPost__avt avt-240")
    for div in divs:
      a = div.find("a")
      link = "https://vietnamnet.vn" + a['href']
      list_link.append(link)
      list_href.append(a['href'])
      #print(list_link)
      #print(a['href'])
    #print(f"length of element {len(divs)}")
    #print(divs[0].prettify())
  else:
    print("Lỗi (response.status_code)")
print(list_link)
```

```python
def get_content(link: str):
    response = requests.get(link)
    if response.status_code == 200:
        html_content = response.text
        soup = BeautifulSoup(html_content, 'html.parser')

        # Title
        title = soup.find('h1', class_="content-detail-title")
        if title:
            list_title.append(title.text.strip())
        else:
            list_title.append("")

        # Summary
        sumary = soup.find('h2', class_=["content-detail-sapo", "content-detail-sapo sm-sapo-mb-0"]) #vì h2 có nhiều class nên dùng list
        if sumary:
            list_sumary.append(sumary.text.strip())
        else:
            list_sumary.append("")
            print(f"@@@@@@@KHÔNG TÌM THẤY TÓM TẮT H2 Ở LINK NÀY@@@@ {link}")

        # Content
        content = soup.find('div', id="maincontent")
        if not content:
            content = soup.find('div', class_="maincontent main-content content-full-image content-full-image-v1")

        if content:
            ps = content.find_all('p')
            insight = ""
            for p in ps:
                insight += p.get_text().strip() + "\n"
            list_insight.append(insight.strip())
        else:
            list_insight.append("")
    else:
        print(f"Lỗi response.status_code {response.status_code} ở link: {link}")
        list_title.append("")
        list_sumary.append("")
        list_insight.append("")


list_title = []
list_insight = []
list_sumary = []
list_content = []

for link in list_link:
  get_content(link)

for i in range(len(list_sumary)):
  full = list_sumary[i] + list_insight[i]
  list_content.append(full)

for i in range(len(list_content)):
  print(list_content[i])
```

```python
import re
from underthesea import sent_tokenize

def cleaning(text):
  text = re.sub(r"[~!@#$%^&*?\/|]", "", text).lower()
  text = re.sub(r"\s+", " ", text).capitalize()
  tokenize = sent_tokenize(text)
  return tokenize
for i in range(len(list_content)):
  cleaned = cleaning(list_content[i])
  print(cleaned)
```

```python
import pandas as pd

data = {
    'title': list_title,
    'content': list_content,
    'url': list_link
}
df = pd.DataFrame(data)
df.to_csv('copus.csv', index=False, encoding='utf-8')

read_csv = pd.read_csv('copus.csv')
#print(read_csv)

# Nối title và content thành một biến doc cho từng bài
docs = read_csv['title'] + ' ' + read_csv['content']


# In ra docs
print(docs[:5])
```

So sánh hai công cụ Web Scraping: BeautifulSoup và Scrapy

Cả hai đều được dùng để lấy dữ liệu từ website, nhưng mỗi cái có cách dùng và ưu nhược điểm khác nhau. Dưới đây là mấy điểm khác biệt dễ thấy nhất:

1. Dễ học, dễ tiếp cận
BeautifulSoup: Cái này rất dễ học, kiểu như chỉ cần biết chút Python là có thể dùng được ngay. Cú pháp đơn giản, dễ hiểu, tra cứu cũng nhanh. Nói chung là hợp cho những ai mới bắt đầu tập tành scrape dữ liệu từ web.

Scrapy: Ngược lại thì Scrapy phức tạp hơn vì nó là cả một framework. Nếu chưa quen hoặc chưa biết nhiều về lập trình, dùng Scrapy có thể hơi choáng vì phải hiểu cách hoạt động, cấu trúc của nó. Nhưng đã hiểu rồi thì dùng rất mạnh.

2. Tốc độ xử lý
Scrapy: Cái này chạy rất nhanh vì nó xử lý theo kiểu bất đồng bộ, tức là có thể gửi nhiều request cùng lúc mà không cần chờ cái này xong rồi mới đến cái kia. Phù hợp với mấy dự án cần lấy hàng trăm, hàng ngàn trang.

BeautifulSoup: Chạy theo kiểu tuần tự, nghĩa là phải chờ từng trang tải xong rồi mới xử lý tiếp. Vì vậy nếu scrape số lượng lớn thì sẽ chậm hơn hẳn. Nếu muốn tăng tốc thì phải kết hợp thêm mấy thư viện khác, hơi rối.

3. Khả năng mở rộng và quản lý
Scrapy: Cái hay của Scrapy là mọi thứ đã được chuẩn bị sẵn như hệ thống quản lý link, lưu dữ liệu, xử lý pipeline, v.v. Tức là làm dự án lớn sẽ dễ hơn, không bị rối. Ngoài ra còn có log, debug, proxy,... đều tích hợp sẵn.

BeautifulSoup: Thì chỉ có nhiệm vụ chính là phân tích HTML thôi. Còn lại từ gửi request, xử lý dữ liệu, lưu file,... đều phải tự viết code. Nếu chỉ làm nhỏ nhỏ thì ổn, còn làm bài bản thì hơi cực.
